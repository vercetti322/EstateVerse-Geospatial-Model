```python
import pandas as pd
df = pd.read_csv('merged_data.csv')
df
```

```python
df.columns
```

```python
# Function to generate random coordinates around a center coordinate
import numpy as np
# Function to generate random coordinates around a center coordinate
def generate_random_data(row, num_points=150, radius=0.015):
    locality_data = {
        'Locality': [row['Locality']] * num_points,
        ' Average Price': np.random.randint(row[' Average Price'] - 500, row[' Average Price'] + 500, num_points),
        'Latitude': np.random.uniform(row['Latitude'] - radius, row['Latitude'] + radius, num_points),
        'Longitude': np.random.uniform(row['Longitude'] - radius, row['Longitude'] + radius, num_points)
    }
    return pd.DataFrame(locality_data)

# Apply the function to each row in the DataFrame
expanded_df = pd.concat([generate_random_data(row) for _, row in df.iterrows()], ignore_index=True)

# Display the resulting DataFrame
expanded_df
```

```python
expanded_df = expanded_df.drop(columns='Locality')
expanded_df
```

```python
expanded_df.to_csv('training_data.csv')
```
