```python
# resolve imports
import numpy as np
import pandas as pd
from route_distance import AmenityDistanceCalculator

superset_amenities = ['parking', 'restaurant', 'fire_station', 'cinema', 'dentist',
       'school', 'atm', 'fuel', 'fast_food', 'toilets', 'pharmacy',
       'bank', 'police', 'post_box', 'place_of_worship', 'bar', 'pub',
       'cafe', 'clinic', 'hospital', 'community_centre',
       'vending_machine', 'bus_station', 'health_post', 'college',
       'theatre', 'showroom', 'marketplace']
```

```python
# Assuming 'Latitude' and 'Longitude' columns are present in new_df
for index, row in new_df.iterrows():
    # Create an instance of AmenityDistanceCalculator for each row
    dist = AmenityDistanceCalculator(row['Latitude'], row['Longitude'], radius=2000)
    
    # Calculate distances
    distances_df = dist.calculate_distances((row['Latitude'], row['Longitude']))
    
    # Map the distances back to the original DataFrame (df)
    for amenity in superset_amenities:
        # Check if the amenity is present in distances_df
        if amenity in distances_df['Amenity'].values:
            # Map the distance to the corresponding row and column
            new_df.at[index, amenity] = distances_df.loc[distances_df['Amenity'] == amenity, 'Amenity_distance_From_point'].values[0]
        else:
            # If amenity is not present, mark as 0
            new_df.at[index, amenity] = 0

# Display the updated DataFrame (df)
new_df
```

```python
import numpy as np
import pandas as pd

# Assuming df is your original DataFrame
df_extrapolated = pd.DataFrame()

def generate_random_values(center, num_points=160, scale=0.015):
    if np.isscalar(center):
        values = center + np.random.exponential(scale, num_points)
    else:
        values = np.array([c + np.random.exponential(scale, num_points) for c in center])
    return values.T  # Transpose the output

# Iterate over each row in the original DataFrame
for _, row in new_df.iterrows():
    # Repeat the 'Locality' column for each set of values
    locality_values = [row['Locality']] * 160

    # Generate random values for the row (excluding 'Locality', 'Latitude', 'Longitude')
    random_values = generate_random_values(row.drop(['Locality', 'Latitude', 'Longitude']))

    # Generate random Latitude and Longitude values within a radius
    latitude_values = generate_random_values(pd.Series([row['Latitude']]), num_points=160)
    longitude_values = generate_random_values(pd.Series([row['Longitude']]), num_points=160)

    # Create a DataFrame for the current row
    expanded_rows = pd.DataFrame(random_values, columns=row.drop(['Locality', 'Latitude', 'Longitude']).index)
    expanded_rows['Locality'] = locality_values
    expanded_rows['Latitude'] = latitude_values
    expanded_rows['Longitude'] = longitude_values

    # Append the current row to the extrapolated DataFrame
    df_extrapolated = pd.concat([df_extrapolated, expanded_rows], ignore_index=True)

# Display the extrapolated DataFrame
print(df_extrapolated)
```
